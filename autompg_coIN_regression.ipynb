```python
pd_data.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']
```

```python
x = pd_data[['weight','cylinders']]
```

```python
y = pd_data[['mpg']]
```

```python
x.shape, y.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x,y)
```

```python
# total data
lr.coef_, lr.intercept_
```
