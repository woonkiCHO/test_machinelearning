```python
import pandas as pd
```

```python
df = pd.read_excel('./files/advertising.xls')
df
```

```python
df.shape
df.info()
```

```python
x = df[['Age', 'Male']]
y = df[['Clicked on Ad']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test,= train_test_split(x, y)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logR = LogisticRegression()
type(logR)
```

```python
logR.fit(x_train, y_train)
```

```python
logR.score(x_train, y_train)
```

```python
logR.classes_
```

```python
logR.coef_
```

```python
pred_train =logR.predict(x_train)
pred_test = logR.predict(x_test)
```

```python
pred_train
```

```python
pred_test
```

```python
logR.predict_proba(x_train)
```

```python
logR.predict_proba(x_test)
```

```python
logR.score(x_train, y_train)
```

```python
from sklearn import metrics
```

```python
metrics.confusion_matrix(y_test, pred_test)
```

```python
print(metrics.classification_report(y_test, pred_test))
```

```python
metrics.confusion_matrix(y_train, pred_train)
```

```python
print(metrics.classification_report(y_train, pred_train))
```
